```python
import pandas as pd 
import numpy as np 
```

```python
movies=pd.read_csv('C:/Users/HP/Downloads/Movies Dataset/movies',sep=',',names=['movie_id','title'])
ratings=pd.read_csv('C:/Users/HP/Downloads/Movies Dataset/u.data',sep='\t',names=['user_id', 'item_id', 'rating', 'timestamp'])
```

```python
#merge
merge_data=pd.merge(movies,ratings,right_on='item_id',left_on='movie_id')
merge_data.head()
```

```python
#association matrix-pivot table
learning_matrix=ratings.pivot_table(index=['item_id'],columns=['user_id'],values='rating').reset_index(drop=True)
learning_matrix.fillna(0,inplace=True)
learning_matrix.head(20)
```

```python
#apply Cosine Similarity Formula on Association Matrix
from scipy.spatial.distance import cosine
#from scipy.spatial.distance import euclidean
#from scipy.spatial.distance import hamming
#from scipy.spatial.distance import minkowski
from sklearn.metrics import pairwise_distances
#Pairwise distance only accept Matrix object that's why we converted it above.
movie_similarity=1-pairwise_distances(learning_matrix,metric='cosine')
np.fill_diagonal(movie_similarity,0)
```

```python
#lets visualize the matrix interms of data frame
ratings_matrix=pd.DataFrame(movie_similarity)
ratings_matrix
```

```python
#lets apply try catch blocks
try:
    user_input=input('Enter movie title to see top 10 similar recommended movies:')
    inp=movies[movies['title']==user_input].index.tolist()
    inp=inp[0]
    #adding similarities column 
    movies['similarity']=ratings_matrix.iloc[inp]
    #printing recommended movies
    print('Recommended movies based on your choice of ',user_input,':\n',movies.sort_values(["similarity"],ascending=False)[1:10])
except:
    print('movie name does not exist in the list')
```
